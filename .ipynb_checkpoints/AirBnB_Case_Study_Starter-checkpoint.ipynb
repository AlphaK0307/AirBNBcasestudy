```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighbourhood groups / which shows up the most
print('5 Neighborhood groups :: Manhattan shows up the most.')
air_bnb.describe(include='all')
air_bnb.groupby(['neighbourhood_group']).sum().sort_values('calculated_host_listings_count')
```

```python
# Are private rooms the most popular in Manhattan?
print('Entire Home/apt is more popular')
air_bnb.describe(include='all')
air_bnb.groupby(['room_type','neighbourhood_group']).count().sort_values('room_type')
```

```python
# Who are the top 10 busiest hosts based on number of reviews per month?

air_bnb.describe(include='all')
air_bnb.groupby(['host_name']).sum().sort_values('number_of_reviews', ascending=False)
```

```python
# Which neighborhood group has the highest avg price and plot the average prices in a bar chart
print('Manhattan has highest avg')
air_bnb.describe(include='all')
air_bnb.groupby('neighbourhood_group').mean()[['price']].sort_values('price')
```

```python
# Which host has the highest minimum rental? *minimal rental is based on price and minimum nights
air_bnb.describe(include='all')
air_bnb.groupby('host_name').mean()[['minimum_nights', 'price']].sort_values('minimum_nights').head()
```

```python
# Who has no (zero) availability with a review count of 100 or more
print('Michael has')
air_bnb.describe(include='all')
air_bnb.groupby(['host_name','availability_365']).sum().sort_values('number_of_reviews').head()
```

```python
# If each host rents all of their properties for the night, which 5 make the most that night?
air_bnb.describe(include='all')
air_bnb.groupby('host_name').sum().sort_values('price', ascending=False).head()
```

```python
# When did Danielle from Queens last receive a review?
air_bnb.describe(include='all')
air_bnb.groupby([['host_name'="Danielle"],'last_review']).sum().sort_values('last_review')
```

```python
# bar chart showing the number of room types in the Bronx
air_bnb.describe(include='all')
air_bnb.groupby(['room_type','neighbourhood_group']).count().sort_values('room_type')
```
